```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pybaselines
import os
from pybaselines import Baseline
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
```

```python
start = 116
end = 225

C_Dopa = ['0','0.05','0.25','0.5','0.75','1','2','3','4','5','7.5','10','20','30']
loc = "D:\\0131\\same baseline"


for i,C in zip(os.listdir(loc), C_Dopa):

    data_url = os.path.join(loc, i)

    df = pd.read_csv(data_url, sep='\s+')
    x = df.iloc[start:end,0]*1E+3
    y = df.iloc[start:end,1]*1E+12
    plt.plot (x, y,':', color='black')
    plt.plot (x, gaussian_filter1d(y,2),label=C+' µM')

    
plt.legend() 
plt.xlabel('mV')
plt.ylabel('pA')
plt.show ()
```
