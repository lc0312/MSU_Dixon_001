```python
import pathlib
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

data_dir = pathlib.Path(r'C:\Users\dixon\Desktop\Nanowire_Code\Nanowire_position')
validation_dir=pathlib.Path(r'C:\Users\dixon\Desktop\Nanowire_Code\Nanowire_position')

image_train= tf.keras.utils.image_dataset_from_directory(
    data_dir,
    labels='inferred',
    label_mode='categorical',
    #class_names=['left', 'no', 'right'],
    color_mode='grayscale',
    batch_size=6,
    image_size=(540,960),
    seed=128,
    validation_split=0.1,
    subset='training',
    )

image_validation= tf.keras.utils.image_dataset_from_directory(
    validation_dir,
    labels='inferred',
    label_mode='categorical',
    #class_names=['left', 'no', 'right'],
    color_mode='grayscale',
    batch_size=6,
    image_size=(540,960),
    seed=128,
    validation_split=0.5,
    subset='validation',
    )


'''
datagen = ImageDataGenerator(rescale=1.0/255)
image_train = datagen.flow_from_directory(
    data_dir,
    target_size=(1080,1920),
    batch_size=6,
    color_mode="grayscale",
    class_mode="sparse",
    shuffle=True,
    subset="training",
    seed=123,
)

vali_datagen = ImageDataGenerator(rescale=1.0/255)

image_validation = vali_datagen.flow_from_directory(
    validation_dir,
    target_size=(1880,1000),
    batch_size=6,
    color_mode="grayscale",
    class_mode="sparse",
    shuffle=False,
    subset="validation",
    seed=123,
)
'''

model = tf.keras.Sequential()
model.add (tf.keras.Input (shape=(540,960)))
#model.add (tf.keras.layers.Reshape ((540,960))) 
model.add (tf.keras.layers.BatchNormalization())
model.add (tf.keras.layers.Bidirectional (tf.keras.layers.GRU (128, return_sequences=True)))
model.add (tf.keras.layers.GRU (64))
model.add (tf.keras.layers.Dense(32,activation='tanh'))
model.add (tf.keras.layers.Dense(12,activation='tanh'))
model.add (tf.keras.layers.Dense(3))

lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    5e-4,
    decay_steps=32,
    decay_rate=0.99,
    staircase=True)

loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True)
opt=tf.keras.optimizers.Adam(learning_rate=lr_schedule)

model.compile (loss=loss, optimizer=opt, metrics=["accuracy"])
model.summary ()

model.fit (image_train, batch_size=12, epochs=10, verbose=2)
model.evaluate (image_validation, batch_size=12, verbose=2)
```
